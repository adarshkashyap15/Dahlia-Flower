```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

**Variable Initialization**

```python
ppr = 12.6
nr = 30
pr = 10
n_p = 140
pf = -1.2
ol = [0.11, 1.1]
```

**Calculation**

```python
pt = (1 / ppr)*np.pi*2
theta = np.linspace(0, n_p * pt, n_p * pr + 1)
```

**Grid generation**

```python
R, THETA = np.meshgrid(np.linspace(0, 1, nr), theta)
R=np.transpose(R)
THETA=np.transpose(THETA)
```

**Mathematical Calculations**

```python
x = 1 - (np.abs(1 - np.mod(ppr * THETA, 2 * np.pi) / np.pi)**2) * 0.7
phi = (np.pi / 2) * np.linspace(ol[0], ol[1], n_p * pr + 1)**2
y = (1.27689*R-1)**2*(R**2)*pf*np.sin(phi)
R2 = x * (R * np.sin(phi) + y * np.cos(phi))
```

**Conversion to Cartesian Coordinates**

```python
X = R2 * np.sin(THETA)
Y = R2 * np.cos(THETA)
Z = x * (R * np.cos(phi) - y * np.sin(phi))
C = np.hypot(np.hypot(X, Y), Z)
```

**Plotting**

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, Z,facecolors=plt.cm.plasma(C),rstride=1,cstride=1,shade=False)
plt.show()
```
